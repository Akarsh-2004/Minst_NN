```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv('./digit-recognizer/train.csv')
```

```python
data = np.array(data)
m, n = data.shape

np.random.shuffle(data)
data = data.T  # Transpose for easier slicing: shape (785, m)
y_all = data[0].astype(int).flatten()  # labels: shape (m,)
x_all = data[1:] / 255.0       # remaining rows = features

# Split into train/dev
x_dev = x_all[:, :1000]
y_dev = y_all[:1000]
x_train = x_all[:, 1000:]
y_train = y_all[1000:]
```

```python
def init_params():
    w1 = np.random.randn(10, 784) * np.sqrt(1. / 784)
    b1 = np.zeros((10, 1))
    w2 = np.random.randn(10, 10) * np.sqrt(1. / 10)
    b2 = np.zeros((10, 1))
    return w1, b1, w2, b2


def ReLU(Z):
    return np.maximum(0, Z)

def ReLU_deriv(Z):
    return Z > 0

def softmax(Z):
    expZ = np.exp(Z - np.max(Z, axis=0, keepdims=True))  # stability fix
    return expZ / np.sum(expZ, axis=0, keepdims=True)

def forward_prop(w1, b1, w2, b2, X):
    Z1 = w1.dot(X) + b1
    A1 = ReLU(Z1)
    Z2 = w2.dot(A1) + b2
    A2 = softmax(Z2)
    return Z1, A1, Z2, A2

def one_hot(Y):
    one_hot_Y = np.zeros((Y.size, Y.max() + 1))
    one_hot_Y[np.arange(Y.size), Y] = 1
    return one_hot_Y.T
```

```python
def backward_prop(Z1, A1, Z2, A2, w2, X, Y):
    m = X.shape[1]
    one_hot_Y = one_hot(Y)
    
    dZ2 = A2 - one_hot_Y
    dw2 = (1 / m) * dZ2.dot(A1.T)
    db2 = (1 / m) * np.sum(dZ2, axis=1, keepdims=True)

    dZ1 = w2.T.dot(dZ2) * ReLU_deriv(Z1)
    dw1 = (1 / m) * dZ1.dot(X.T)
    db1 = (1 / m) * np.sum(dZ1, axis=1, keepdims=True)

    return dw1, db1, dw2, db2
```

```python
def update_params(w1, b1, w2, b2, dw1, db1, dw2, db2, learning_rate):
    w1 -= learning_rate * dw1
    b1 -= learning_rate * db1
    w2 -= learning_rate * dw2
    b2 -= learning_rate * db2
    return w1, b1, w2, b2
```

```python
def get_predictions(A2):
    return np.argmax(A2, axis=0)

def get_accuracy(predictions, Y):
    return np.mean(predictions == Y)
```

```python
def train(X, Y, iterations, learning_rate):
    w1, b1, w2, b2 = init_params()

    for i in range(iterations):
        Z1, A1, Z2, A2 = forward_prop(w1, b1, w2, b2, X)
        dw1, db1, dw2, db2 = backward_prop(Z1, A1, Z2, A2, w2, X, Y)
        w1, b1, w2, b2 = update_params(w1, b1, w2, b2, dw1, db1, dw2, db2, learning_rate)

        if i % 100 == 0:
            predictions = get_predictions(A2)
            acc = get_accuracy(predictions, Y)
            print(f"Iteration {i}: Accuracy = {acc:.4f}")

    return w1, b1, w2, b2
```

```python
w1, b1, w2, b2 = train(x_train, y_train, iterations=1000, learning_rate=0.1)
```

```python
data = data.T  # shape: (785, 43000)
y_all = data[0].astype(int).flatten()        # shape: (43000,)
x_all = data[1:] / 255.0                     # shape: (784, 43000)

# Split into train/dev
x_test = x_all[:, 1000:]     # shape: (784, 42000)
y_test = y_all[1000:]        # shape: (42000,)


Z1_test = w1.dot(x_test) + b1
A1_test = ReLU(Z1_test)
Z2_test = w2.dot(A1_test) + b2
A2_test = softmax(Z2_test)

predictions = np.argmax(A2_test, axis=0)
accuracy_test = np.mean(predictions == y_test)
print("Test Accuracy:", accuracy_test)
```
